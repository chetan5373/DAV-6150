# 1. Introduction

- The way healthcare providers prescribe medications under the Medicare Part D program has a huge impact on patient care and costs. Medicare Part D covers prescription drugs for over 47 million people, mainly seniors and those with disabilities. With prescription costs expected to surpass USD1 trillion by 2024, its important to understand how doctors choose which medications to prescribe.

- One major concern is the overprescribing of opioids. In 2021, about 25% of all Medicare Part D prescriptions were for opioids, with some doctors writing 300 prescriptions a month. This has contributed to the opioid crisis, leading to over 81,000 overdose deaths in one year.

- Another issue is the unnecessary prescribing of antibiotics. Research shows that nearly 30% of antibiotic prescriptions are not needed, which can lead to antibiotic resistance. This resistance is responsible for about 23,000 deaths in the U.S. each year.

- Analyzing how providers prescribe medications is crucial to improving patient care, reducing costs, and ensuring safer prescribing practices.

# 2. Research Questions

__Real World Applications for each Research Question__

- Q.1) How do total drug costs for Medicare beneficiaries aged 65 and older compare between generic and brand-name medications?
  - Applications: Understanding this can promote the use of generics, reducing costs for beneficiaries and the Medicare program while improving medication accessibility.

- Q.2) What is the relationship between the number of claims and the total drug costs for generic medications among beneficiaries aged 65 and older?
  - Applications: Analyzing this relationship can help manage spending and educate beneficiaries on the benefits of generics, potentially lowering overall costs.

- Q.3) How do prescription rates for opioids and antibiotics vary across different states?
  - Applications: Identifying state-level variations can guide targeted public health initiatives to reduce inappropriate prescribing and address opioid misuse.

- Q.4) What factors influence the total cost of prescriptions under Medicare Part D?
  - Applications: Insights into these factors can inform policies to control costs and enhance drug pricing negotiations, benefiting beneficiaries financially.

- Q.5) How does the number of beneficiaries correlate with the total claims and costs for various drug categories?
  - Applications: Examining this correlation can improve resource allocation and tailor healthcare services to meet the specific needs of different populations.

# 4. Approach

#### Data Management:

- We will download the dataset directly from the CMS website and upload it to GitHub. Then, we will access the data using the GitHub link with Pandas in Python for cleaning, preprocessing, and analysis.

#### Statistical Analysis:

- We will use the following statistical techniques:

  - Descriptive Statistics: To summarize the data and identify trends in prescriptions.
  - Correlation Analysis: To see if there are relationships between different factors, such as opioid prescriptions and overdose rates.
  - Regression Analysis: To find out what influences prescribing habits and predict future outcomes.

#### Visualizations:

- We will create several types of visualizations, including:

  - Bar Charts
  - Heatmaps
  - Distplot
  - Boxplot
  - Line Graphs
  
  
#### Model Construction:
- We will build three types of models:

  - Decision Tree: For easy-to-understand rules about prescribing patterns.
  - XGBoost: For accurate predictions using advanced techniques.
  - Random Forest: An ensemble of decision trees to improve prediction reliability.
  
#### To compare the models, we will look at:

- Accuracy: The overall percentage of correct predictions.
- Precision: The true positive predictions out of all positive predictions.
- Recall: The true positive predictions out of all actual positives.
- F1 Score: A balance between precision and recall.
- AIC: A measure used to compare models, taking into account the number of parameters. Lower values indicate a better model fit.
- BIC : Similar to AIC, BIC also penalizes for the number of parameters in the model, with a stronger penalty for more complex models.
- ROC AUC: A performance measurement for classification models; it shows the trade-off between true positive rates and false positive rates. Higher values indicate better model performance.

#### Ensemble Model:

- We will combine the predictions from the individual models using a stacking ensemble approach. The predictions from each model will be used as inputs for a final model (like KNN, logistic regression, etc.) to improve accuracy.
